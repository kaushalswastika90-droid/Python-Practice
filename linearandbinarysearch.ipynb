```python
'''Problem 1 :  You are given a list of strings representing the name of students in a class and their corresponding test scores.
Implement a function to perform linear search to find the test score of a specific student.
If the student is found in the list, return their test score.
If the student is not found, return "Student not found".
Create a Python function named find_test_score that takes two parameters:
student_scores: a list of tuples where each tuple contains a student's name (string) and their corresponding test score (integer).
student_name: a string representing the name of the student whose test score is to be found.
Implement the find_test_score function to perform linear search on the list student_scores to find the test score of the student with the given name.
If the student is found, return their test score.
Otherwise, return "Student not found". '''
def find_test_score(student_scores, student_name):
    for name, score in student_scores:
        if name == student_name:
            return score
    return "Student not found"



student_scores = [
    ("Alice", 85),
    ("Bob", 92),
    ("Charlie", 78),
    ("David", 90)
]

print(find_test_score(student_scores, "Bob"))
print(find_test_score(student_scores, "Charlie"))
print(find_test_score(student_scores, "Eve"))
```

```python
'''Problem 2 : ( Medium ) : You are given a list of dictionaries representing books in a library.
Each dictionary contains information about a book, including its title, author, genre, and publication year.
Implement a function to perform linear search to find books published in a specific year.
If books are found, return a list containing titles of books published in that year. If no books are found, return an empty list.
Create a Python function named find_books_by_year that takes two parameters:
Library : a list of dictionaries where each dictionary contains information about a book.
Year : an integer representing the publication year to search for.
Implement the find_books_by_year function to perform linear search on the list library to find books published in the specified year.
Return a list containing titles of books published in that year, or an empty list if no books are found.'''
def find_books_by_year(library, year):
  # Write your code here
  b=[]
  for book in library:
    if book['publication_year'] == year:
      b.append(book['title'])
  return b


# Example use:
library = [ {'title': 'Book1', 'author': 'Author1', 'genre': 'Fiction', 'publication_year': 2000},
 {'title': 'Book2', 'author': 'Author2', 'genre': 'Non-Fiction', 'publication_year': 2010},
  {'title': 'Book3', 'author': 'Author3', 'genre': 'Science Fiction', 'publication_year': 2020},
   {'title': 'Book4', 'author': 'Author3', 'genre': 'Science Fiction', 'publication_year': 2020},
    {'title': 'Book5', 'author': 'Author5', 'genre': 'Mystery', 'publication_year': 2001} ]
year_to_search = 2020
print("Books published in", year_to_search, ":", find_books_by_year(library, year_to_search))
```

```python
'''Problem 3 : ( Hard ) : You are given a list of lists where each inner list represents a row in a matrix.
Implement a function to perform linear search to find the position of a specific element in the matrix.
If the element is found, return its row and column indices. If the element is not found, return (-1, -1).
Create a Python function named find_element_in_matrix that takes two parameters:
matrix: a list of lists representing a matrix.
element: an integer representing the element to search for.
Implement the find_element_in_matrix function to perform linear search on the matrix to find the position of the specified element.
Return a tuple containing its row and column indices if the element is found, or (-1, -1) if the element is not found. '''
def find_element_in_matrix(matrix, element):
  # Write your code here
  for i in range (len(matrix)):
    for j in range (len(matrix[i])):
      if matrix[i][j]==element:
        return i,j
  return -1,-1


# Example usage:
matrix = [     [1, 2, 3],     [4, 5, 6],     [7, 8, 9] ]
element_to_search = 3
print("Position of", element_to_search, ":", find_element_in_matrix(matrix, element_to_search))
```

```python
'''Problem 1 :  You are given a sorted list of integers and a target value.
Implement a function to perform binary search to find the index of the closest element to the target value in the list.
If there are multiple closest elements, return the index of the smallest one.'''
def binary_search_closest(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    # If target is smaller than the first element
    if left == 0:
        return 0

    # If target is greater than the last element
    if left == len(arr):
        return len(arr) - 1

    # Compare the two closest elements
    if target - arr[left - 1] <= arr[left] - target:
        return left - 1
    else:
        return left


# Example usage
my_list = [1, 3, 5, 7, 9, 11, 13]
target_value = 6

print("Index of closest element to", target_value, ":", binary_search_closest(my_list, target_value))
```
