```python
'''1. Safe Division Counter
Problem Statement: You are given two integer arrays A and B of equal length. For each index i, divide A[i] by B[i].
If division by zero occurs, ignore that index and continue execution.'''
n=int(input())
a=list(map(int,input().split()))
b=list(map(int,input().split()))
sum=0
for i in range(n):
  try:
    sum+=a[i]//b[i]
  except ZeroDivisionError:
    pass
print(a)
print(b)
print(sum)
```

```python
'''3. ATM Withdrawal Simulator
Problem Statement: Simulate ATM withdrawal using exception handling.
Withdrawal is valid only if amount is a multiple of 100 and less than or equal to balance.'''
try:
  balance=int(input())
  withdrawal=int(input())

  if withdrawal%100!=0 or withdrawal>balance:
    raise ValueError
  else:
    print("Transaction done",balance-withdrawal)
except ValueError:
  print("Transaction Failed")
```

```python
'''5. Average Marks Calculator
Problem Statement: Calculate average marks ignoring invalid values (negative or >100).
Input Format: n n integers Output Format: Average of valid marks '''
n=int(input())
total=0
count=0
for i in range(n):
  try:
    mark=int(input())
    if mark<0 or mark>100:
      raise ValueError
    else:
      total+=mark
      count+=1
  except ValueError:
    pass
print(total/count)
```

```python
'''7. Password Strength Checker
Problem Statement: A password is strong if length ≥ 8 and contains at least one digit.
'''
n = int(input())

count = 0

for i in range(n):
  try:
    p = input()

    if len(p) >= 8 and any(ch.isdigit() for ch in p):
        count += 1
    else:
      raise ValueError
  except ValueError:
    pass

print("No. of strong pass: ",count)
```

```python
'''9. Calculator with Exception Rules
Problem Statement: Perform arithmetic operations. Handle invalid operators and division by zero.'''
n = int(input())

for i in range(n):
    try:
        a, op, b = input().split()
        a = int(a)
        b = int(b)

        if op == "+":
            print(a + b)
        elif op == "-":
            print(a - b)
        elif op == "*":
            print(a * b)
        elif op == "/":
            print(a // b)
        else:
            raise ValueError

    except (ValueError, ZeroDivisionError):
        print("Invalid")
```

```python
'''10. Login Attempt Tracker
Problem Statement: Track login attempts. Lock account after three failed attempts.'''
n = int(input())

correct = "admin123"
fail = 0

for i in range(n):
  try:
    p = input()

    if p == correct:
        print("Login Successful")
        break

    fail += 1

    if fail == 3:
        raise ValueError

  except ValueError:
    print("Account Locked")
```
