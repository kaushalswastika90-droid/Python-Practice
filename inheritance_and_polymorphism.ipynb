```python
'''Problem 1: Single Inheritance – Person and Student
Problem Statement:
Create a Person class with attributes name and age. Inherit a Student class from Person with an additional attribute roll_no.
Print the details of a student.
Input Format: First line: Name (string),
Second line: Age (integer),
Third line: Roll number (integer)
Output Format: Print name, age, and roll number in separate lines.
Sample Input: Alice | 20 | 101
Sample Output: Name: Alice | Age: 20 | Roll No: 101'''
class Person:
  def __init__(self,name,age):
    self.name=name
    self.age=age

class Student(Person):
  def __init__(self,name,age,roll_no):
    Person.__init__(self,name,age)
    self.roll_no=roll_no

name=input()
age=int(input())
roll_no=int(input())
s=Student(name,age,roll_no)
print("Name:",s.name,"| Age:",s.age,"| Roll No:",s.roll_no)1
```

```python
'''Problem 2: Multilevel Inheritance – Vehicle and Car
Problem Statement:
Create a Vehicle class with attribute brand. Create a Car class inheriting Vehicle with attribute model. Create a SportsCar class inheriting
Car with attribute top_speed. Print details of a SportsCar.
Input Format: First line: brand (string),
Second line: model (string),
Third line: top speed (integer)
Output Format: Print brand, model, and top speed.
Sample Input: Ferrari | F8 | 340
Sample Output: Brand: Ferrari | Model: F8 | Top Speed: 340'''
class Vehicle:
  def __init__(self,brand):
    self.brand=brand

class Car(Vehicle):
  def __init__(self,brand,model):
    Vehicle.__init__(self,brand)
    self.model=model

class Sportscar(Car):
  def __init__(self,brand,model,top_speed):
    Car.__init__(self,brand,model)
    self.top_speed=top_speed

brand=input()
model=input()
top_speed=int(input())
s=Sportscar(brand,model,top_speed)
print("Brand: ",s.brand,"| Model: ",s.model,"| Top Speed: ",s.top_speed)
```

```python
'''Problem 3: Hierarchical Inheritance – Teacher and Student
Problem Statement:
Create a base class Person with attribute name. Derive two classes: Teacher with attribute subject,
Student with attribute grade. Print details.
Input Format: Teacher name, subject, student name, grade
Output Format: Teacher details and student details
Sample Input: Mr. Smith | Math | Alice | A
Sample Output: Teacher Name: Mr. Smith, Subject: Math | Student Name: Alice, Grade: A'''
class Person:
  def __init__(self,name):
    self.name=name

class Teacher(Person):
  def __init__(self,name,subject):
    Person.__init__(self,name)
    self.subject=subject

class Student(Person):
  def __init__(self,name,grade):
    Person.__init__(self,name)
    self.grade=grade

tname=input()
subject=input()
sname=input()
grade=input()
t=Teacher(tname,subject)
s=Student(sname,grade)
print("Teacher Name: ",t.name,", Subject: ",t.subject,"| MAStudent Name: ",s.name,", Grade: ",s.grade)
```

```python
'''Problem 4: Inheritance with Method Overriding
Problem Statement: Create a Shape class with method area() returning 0.
Derive Rectangle class with length and width, override area() to return rectangle area.
Input Format: Two integers length and width
Output Format: Print area Sample Input: 5 | 4
Sample Output: 20'''
class Shape:
  def area(self):
    return 0

class Rectangle(Shape):
  def __init__(self,length,width):
    self.length=length
    self.width=width
  def area(self):
    return self.length*self.width

l=int(input())
b=int(input())
r=Rectangle(l,b)
print("Output: ",r.area())
```

```python
'''Problem 5: Super Keyword in Inheritance
Problem Statement: Create Person class with constructor for name.
Create Employee class inheriting Person with constructor for salary using super().
Print name and salary. Input Format: name, salary
Output Format: Print name and salary
Sample Input: Alice | 50000
Sample Output: Name: Alice | Salary: 50000'''
class Person:
  def __init__(self,name):
    self.name=name

class Employee(Person):
  def __init__(self,name,salary):
    super().__init__(name)
    self.salary=salary

name=input()
salary=int(input())
e=Employee(name,salary)
print("Name:",e.name,"| Salary:",e.salary)
```

```python
'''Problem 6: Function Overriding
Problem Statement: Create Animal base class with sound().
Derive Dog and Cat classes overriding sound().
Input Format: Animal type
Output Format: Print sound
Sample Input: Dog
Sample Output: Bark'''
class Animal:
  def sound(self):
    print("Animal sound")

class Dog(Animal):
  def sound(self):
    print("Bark")

class Cat(Animal):
  def sound(self):
    print("Meow")

animal=input()
if animal == "Dog":
    d = Dog()
    d.sound()
elif animal == "Cat":
    c = Cat()
    c.sound()
```

```python
'''Problem 7: Duck Typing
Problem Statement: Write Car and Bike classes both with wheels() method.
Create function display_vehicle(obj).
Input Format: Car or Bike
Output Format: Print wheels info
Sample Input: Bike
Sample Output: Two Wheels'''
class Car:
  def wheels(self):
    print("Four Wheels")

class Bike:
  def wheels(self):
    print("Two Wheels")

def display_vehicle(obj):
  obj.wheels()

vehicle = input()

if vehicle == "Car":
    c = Car()
    display_vehicle(c)
elif vehicle == "Bike":
    b = Bike()
    display_vehicle(b)
```

```python
'''Problem 8: Operator Overloading
Problem Statement: Create Vector class with x, y. Overload + to add vectors.
Input Format: x1 y1 | x2 y2
Output Format: (x, y)
Sample Input: 2 3 | 4 5
Sample Output: (6, 8)'''
class Vector:
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def __add__(self,other):
    return Vector(self.x+other.x,self.y+other.y)

v1 = Vector(2, 3)
v2 = Vector(4, 5)

v3 = v1 + v2

print(v3.x, v3.y)
```

```python
'''Problem 9: Method Overloading Simulation
Problem Statement: Create Calculator class with add() method using default args to accept 2 or 3 numbers.
Input Format: count n1 n2 (n3)
Output Format: Sum
Sample Input: 3 | 5 7 9
Sample Output: 21'''
class Calculator:
    def add(self, a, b, c=0):
        return a + b + c

cal = Calculator()

count = int(input())

if count == 2:
    n1, n2 = map(int, input().split())
    print(cal.add(n1, n2))
elif count == 3:
    n1, n2, n3 = map(int, input().split())
    print(cal.add(n1, n2, n3))
```

```python
'''Problem 10: Abstract Class Polymorphism
Problem Statement: Create abstract Shape with area().
Implement Rectangle and Circle. Input Format: Shape type and parameters
Output Format: Print area
Sample Input: Rectangle | 5 | 4
Sample Output: 20
Sample Input: Circle | 7
Sample Output: 153'''
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 22 / 7 * self.radius * self.radius

shape = input()

if shape == "Rectangle":
    length = int(input())
    width = int(input())
    r = Rectangle(length, width)
    print(int(r.area()))

elif shape == "Circle":
    radius = int(input())
    c = Circle(radius)
    print(int(c.area()))
```
