```python
'''
1. Amazon - Warehouse Stock Reordering
Problem Statement:
Given an array of integers representing stock of each product, return indices where stock < threshold.
Input Format:
n
t
n space-separated integers
Output:
Indices needing restocking.
Example:
Input:
6
10
12 3 9 4 11 2
Output:
1 2 3 5
'''

def restock_indices(stock, threshold):
  #Write your solution code here
  result = []
  for i in range(len(stock)):
    if stock[i]<threshold:
      result.append(i)
  return result

n = int(input())
t = int(input())
arr = list(map(int, input().split()))

result = restock_indices(arr, t)
print(*result)
```

```python
'''
2. Google – Max Continuous Working Hours
Problem Statement:
Find maximum consecutive days with hours ≥ limit
Input:
n
limit
hours[]
Output:
Max consecutive days.
Example:
Input:
7
6
8 5 7 6 6 4 9
Output:
3
'''
def max_consecutive_days(hours, limit):
   #Write your solution code here
   count=0
   max_count=0
   for h in hours:
    if h>=limit:
      count+=1
    else:
      count=0
    max_count=max(max_count,count)
   return max_count

n = int(input())
limit = int(input())
arr = list(map(int, input().split()))
print(max_consecutive_days(arr, limit))
```

```python
'''
4. Flipkart – Price Correction System
Problem:
Replace product prices > max_price with max_price.
Input:
n
max_price
prices[]
Output:
Corrected prices.
Example:
Input:
5
1000
900 1200 500 2000 800
Output:
900 1000 500 1000 800
'''
def correct_prices(prices, max_price):
    #Write your solution code here
  for i in range(len(prices)):
    if prices[i]>max_price:
      prices[i]=max_price
  return prices

n = int(input())
max_price = int(input())
arr = list(map(int, input().split()))

result = correct_prices(arr, max_price)
print(*result)
```

```python
'''
5. Netflix - User Watch Pattern
Problem:
Find longest strictly increasing streak.
Input:
n
array[]
Output:
Length.
Example:
Input:
7
30 40 35 50 60 20 25
Output:
3
'''
def longest_increasing_streak(arr):
  count=0
  max_count=0
    #Write your solution code here
  for i in range(len(arr)-1):
    if arr[i]<arr[i+1]:
      count+=1
    else:
      count=0
    max_count=max(max_count,count)
  return max_count+1

n = int(input())
arr = list(map(int, input().split()))

print(longest_increasing_streak(arr))
```

```python
'''
6. Ola – Trip Fare Outlier Detection
Problem:
Find fares greater than mean + 2*std.
Input:
n
fares[]
Output:
Outliers.
Example:
Input:
6
200 220 230 1000 240 250
Output:
1000
'''
import math

def find_outliers(arr):
  #Write solution code here
  mean=sum(arr)/len(arr)

  variance=0
  for x in arr:
    variance+=(x-mean)**2
  variance=variance/len(arr)

  sd=math.sqrt(variance)

  limit=mean+2*sd

  result=[]

  for x in arr:
    if x>limit:
      result.append(x)
  return result

n = int(input())
arr = list(map(int, input().split()))

result = find_outliers(arr)
print(*result)
```

```python
'''
8. Apple – Temperature Sensor Peak Detection
Problem:
Return all local peaks.
Input:
n
temps[]
Output:
Peaks.
Example:
Input:
6
3 7 4 9 8 6
Output:
7 9
'''
def local_peaks(arr):
  result=[]
  #write solution code here
  for i in range(1,len(arr)-1):
    if arr[i]>arr[i-1] and arr[i]>arr[i+1]:
      result.append(arr[i])
  return result


n = int(input())
arr = list(map(int, input().split()))

result = local_peaks(arr)
print(*result)
```

```python
'''
9. Uber – Surge Pricing Predictor
Problem:
Return first index where multiplier > k.
Input:
n
k
array[]
Output:
Index or -1.
Example:
Input:
5
1.5
1.2 1.3 1.5 1.8 1.4
Output:
3
'''
def surge_index(arr, k):
  #Write solution code here
  for i in range(len(arr)):
        if arr[i] > k:
            return i
  return -1

n = int(input())
k = float(input())
arr = list(map(float, input().split()))

print(surge_index(arr, k))
```
