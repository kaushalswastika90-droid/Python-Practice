```python
# Problem 1: Grocery Shelf Arrangement (Insertion Sort Based)

# Problem Statement:
# You are managing a grocery store shelf where items arrive one by one.
# Each item has a freshness score.
# Using insertion-sort style shifting, insert each incoming item so the shelf
# remains sorted in decreasing order.

# Input:
# n
# n space-separated freshness scores

# Output:
# Final arrangement after insertion-sort simulation.

# Example:
# Input:
# 6
# 30 10 50 40 20 60
#
# Output:
# 60 50 40 30 20 10
n=int(input())
arr=list(map(int,input().split()))
def insertionSort(arr):
  for i in range(1,n):
    key=arr[i]
    j=i-1

    while j>=0 and key>arr[j]:
      arr[j+1]=arr[j]
      j-=1
    arr[j+1]=key
  return arr

b=insertionSort(arr)
print(*b)
```

```python
# Problem 2: Hiring the Best Candidate (Selection Sort Based)

# Problem Statement:
# A company selects candidates by repeatedly picking the highest score first,
# mimicking selection sort.
# Return the final sorted list using minimal swaps.

# Input:
# n
# n space-separated scores

# Output:
# Scores sorted in descending order using selection-sort logic.

# Example:
# Input:
# 5
# 12 90 56 23 80
#
# Output:
# 90 80 56 23 12
n=int(input())
arr=list(map(int,input().split()))
def ss(arr):
  for i in range(n-1):
    min_index=i
    for j in range(i+1,n):
      if arr[j]>arr[min_index]:
        min_index=j
    arr[i],arr[min_index]=arr[min_index],arr[i]
  return arr

b=ss(arr)

print(*b)
```

```python
# Problem 3: Temperature Stabilization in a Greenhouse (Bubble Sort Based)

# Problem Statement:
# Temperatures fluctuate in a greenhouse.
# The system stabilizes them using bubble-sort style adjacent comparisons
# and swaps.
# Return the temperature list after complete bubble sorting.

# Input:
# n
# n space-separated temperatures

# Output:
# Temperatures sorted in increasing order using Bubble Sort.

# Example:
# Input:
# 7
# 32 28 35 30 25 40 20
#
# Output:
# 20 25 28 30 32 35 40
a=int(input())
arr=list(map(int,input().split()))
n=len(arr)

def bb(arr):
  for i in range(n):
    for j in range(n-i-1):
      if arr[j]>arr[j+1]:
        arr[j],arr[j+1]=arr[j+1],arr[j]
  return arr
b=bb(arr)
print(*b)
```
