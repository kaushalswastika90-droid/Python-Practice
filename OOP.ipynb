```python
'''1. Bank Account Balance (Encapsulation)
Problem Statement: Create a class BankAccount with a private variable __balance. Implement
methods to deposit(amount) and withdraw(amount). Withdrawal should not be allowed if balance is
insufficient.
Input Format:
- First line: Initial balance (integer)
- Second line: Number of operations n
- Next n lines: Operation type and amount (deposit 100 or withdraw 50)
Output Format: Print the final balance after all operations.
Sample Input:
500
3
deposit 200
withdraw 100
withdraw 700
Sample Output:
600'''

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient balance!")

    def get_balance(self):
        return self.__balance


balance = int(input("Enter initial balance: "))
n = int(input("Enter number of operations: "))

account = BankAccount(balance)

for i in range(n):
    operation, amount = input().split()

    amount = int(amount)

    if operation == "deposit":
        account.deposit(amount)
    elif operation == "withdraw":
        account.withdraw(amount)

print("\nFinal Balance:", account.get_balance())
```

```python
"""2. Student Marks (Encapsulation with Getter/Setter)
Problem Statement: Create a class Student with private variable __marks.
Provide setter and getter methods.
Ensure that marks cannot be set to negative values (if negative, set as 0).
Input Format:
- First line: Student name
- Second line: Marks (integer)
Output Format: Print student name and valid marks.
Sample Input: Rahul-25
Sample Output: Rahul 0
"""
class Student:
    def __init__(self, name):
        self.name = name
        self.__marks = 0

    def set_marks(self, marks):
        if marks < 0:
            self.__marks = 0
        else:
            self.__marks = marks

    def get_marks(self):
        return self.__marks

name = input()
marks = int(input())

s = Student(name)
s.set_marks(marks)

print(s.name, s.get_marks())
```

```python
"""3. ATM Machine (Encapsulation)
Problem Statement: Create a class ATM with private variable __pin.
The class should allow updating the pin only through a method change_pin(old, new).
Change should succeed only if old pin is correct.
Input Format:
- First line: Initial pin
- Second line: Old pin entered by user
- Third line: New pin to set
Output Format: Print 'PIN changed successfully' if updated, else 'Incorrect Old PIN'.
Sample Input: 1234 1234 9876
Sample Output: PIN changed successfully"""
class ATM:
    def __init__(self, pin):
        self.__pin = pin

    def change_pin(self, old, new):
        if old == self.__pin:
            self.__pin = new
            print("PIN changed successfully")
        else:
            print("Incorrect Old PIN")


# Input
pin = int(input())
old = int(input())
new = int(input())

atm = ATM(pin)
atm.change_pin(old, new)
```

```python
"""4. Employee Salary Slip (Encapsulation)
Problem Statement: Create a class Employee with private attributes __name, __salary.
Salary must not be accessible directly.
Provide method get_salary() to display it only.
Input Format:
- First line: Employee name
- Second line: Salary
Output Format: Print 'Employee: , Salary: '.
Sample Input: Amit 50000
Sample Output: Employee: Amit, Salary: 50000"""
class Employee:
    def __init__(self, name, salary):
        self.__name = name
        self.__salary = salary

    def get_salary(self):
        print(f"Employee: {self.__name}, Salary: {self.__salary}")


# Input
name = input()
salary = int(input())

e = Employee(name, salary)
e.get_salary()
```

```python
"""5. Abstract Class Shape (Abstraction - abc module)
Problem Statement: Create an abstract class Shape with abstract method area().
Implement subclasses Circle and Rectangle.
Input Format:
- First line: Shape type (circle or rectangle)
- If circle: provide radius
- If rectangle: provide length and breadth
Output Format: Print the area of the given shape (rounded to 2 decimals).
Sample Input:
circle
7
Sample Output:
153.94"""

from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, r):
        self.r = r

    def area(self):
        return math.pi * self.r * self.r

class Rectangle(Shape):
    def __init__(self, l, b):
        self.l = l
        self.b = b

    def area(self):
        return self.l * self.b


# Input
shape = input()

if shape == "circle":
    r = int(input())
    obj = Circle(r)
elif shape == "rectangle":
    l, b = map(int, input().split())
    obj = Rectangle(l, b)

print(f"{obj.area():.2f}")
```

```python
"""6. Payment System (Abstraction)
Problem Statement: Create an abstract class Payment with abstract method pay(amount).
Implement subclasses CreditCardPayment and UPIPayment.
Input Format:
- First line: Payment method (credit or upi)
- Second line: Amount
Output Format: Print 'Paid using '.
Sample Input:
upi
250
Sample Output:
Paid 250 using UPI"""
from abc import ABC,abstractmethod
class Payment(ABC):
  @abstractmethod
  def pay(self,amount):
    pass

class CreditCardPayment(Payment):
    def pay(self,amount):
      print(f"Paid {amount} using Credit Card")

class UPIPayment(Payment):
    def pay(self,amount):
      print(f"Paid {amount} using UPI")

mode = input()

if mode == "CreditCardPayment":
    amount=int(input())
    obj = CreditCardPayment()

elif mode == "UPIPayment":
    amount=int(input())
    obj = UPIPayment()

obj.pay(amount)
```

```python
"""7. Vehicle Speed (Abstraction + Encapsulation)
Problem Statement: Create an abstract class Vehicle with abstract method max_speed().
Create two subclasses Car and Bike.
Each should store a private variable __speed.
Input Format:
- First line: Vehicle type (car or bike)
- Second line: Speed
Output Format: Print 'Max speed of is '.
Sample Input:
car
180
Sample Output:
Max speed of car is 180"""

from abc import ABC,abstractmethod
class Vehicle(ABC):
  @abstractmethod
  def max_speed(self):
    pass

class Car(Vehicle):
  def __init__(self,speed):
    self.__speed=speed

  def max_speed(self):
    print(f"Max speed of Car is {self.__speed}")

class Bike(Vehicle):
  def __init__(self,speed):
    self.__speed=speed

  def max_speed(self):
    print(f"Max speed ofBike  is {self.__speed}")

vehicle=input()
if vehicle=="Car":
  speed=int(input())
  obj=Car(speed)
elif vehicle=="Bike":
  speed=int(input())
  obj=Bike(speed)

obj.max_speed()
```

```python
"""8. Online Exam System (Encapsulation)
Problem Statement: Create a class Exam with private attributes __username and __password.
Provide login method that checks credentials.
Input Format:
- First line: Username
- Second line: Password
Output Format: Print 'Login Successful' if username is 'admin' and password is '1234', else 'Login Failed'.
Sample Input: admin 1234
Sample Output: Login Successful"""

class Exam:
  def __init__(self,username,password):
    self.username=username
    self.password=password

  def login(self):
    if self.username=="admin" and self.password=="1234":
      print("Login Successful")
    else:
      print("Login Failed")

username=input()
password=input()
obj=Exam(username,password)
obj.login()
```

```python
"""9. Abstract Animal Sounds (Abstraction)
Problem Statement: Create an abstract class Animal with abstract method make_sound().
Implement subclasses Dog and Cat. Input Format:- First line: Animal type (dog or cat)
Output Format: Print the sound of that animal. (Dog -> Woof, Cat -> Meow)
Sample Input: cat
Sample Output: Meow"""

from abc import ABC,abstractmethod

class Animal(ABC):
  @abstractmethod
  def make_sound(self):
    pass

class Dog(Animal):
  def make_sound(self):
    print("Woof")

class Cat(Animal):
  def make_sound(self):
    print("Meow")

animal=input()
if animal=="Dog":
  obj=Dog()
elif animal=="Cat":
  obj=Cat()

obj.make_sound()
```

```python
"""10. Encapsulation in Library Book
Problem Statement: Create a class Book with private attributes __title and __author.
Provide methods set_details(title, author) and get_details().
Input Format:
- First line: Title
- Second line: Author
Output Format: Print 'Book: by '.
Sample Input: Python Programming Guido van Rossum
Sample Output: Book: Python Programming by Guido van Rossum"""

class Book:
  def __init__(self,title,author):
    self.__title=title
    self.__author=author

  def set_details(self,title,author):
    self.__title=title
    self.__author=author

  def get_details(self):
    print(f"Book: {self.__title} by {self.__author}")

title = input()
author = input()

obj = Book(title, author)
obj.get_details()
```

```python
#single inheritence
class parent:
  def func1(self):
    print("Hello Parent")

class child(parent):
  def func2(self):
    print("Hello Child")

test=child()
test.func1()
test.func2()
```

```python
#Multiple Inheritance
class parent1:
  def func1(self):
    print("Hello Parent 1")

class parent2:
  def func2(self):
    print("Hello Parent 2")

class parent3:
  def func3(self):
    print("Hello Parent 3")

class child(parent1,parent2,parent3):
  def func4(self):
    print("Hello Child")

test=child()
test.func1()
test.func2()
test.func3()
test.func4()
```

```python
#multileevel inheritence
class grandparent:
  def func1(self):
    print("Hello Grandparent")

class parent(grandparent):
  def func2(self):
    print("Hello Parent")

class child(parent):
  def func3(self):
    print("Hello Child")

test=child()
test.func1()
test.func2()
test.func3()
```

```python
#hierarchical inheritence
class parent:
  def func1(self):
    print("Hello Parent")

class child1(parent):
  def func2(self):
    print("Hello Child1")

class child2(parent):
  def func3(self):
    print("Hello Child2")

test1=child1()
test2=child2()
test1.func1()
test1.func2()
test2.func1()
test2.func3()
```

```python

# Hybrid Inheritance

class parent:
    def func1(self):
        print("Hello Parent")

class child1(parent):
    def func2(self):
        print("Hello Child1")

class child2(parent):
    def func3(self):
        print("Hello Child2")

class child3(child1, child2):
    def func4(self):
        print("Hello Child3")

test = child3()
test.func1()
test.func2()
test.func3()
test.func4()
```
