```python
# Problem 1: Bubble Sort an Array
# Problem Statement:
# Sort an array of integers in ascending order using Bubble Sort.
#
# Input:
# n
# array
#
# Output:
# Sorted array

arr=list(map(int,input().split()))
for i in range(len(arr)):
  for j in range(len(arr)-i-1):
    if arr[j]>arr[j+1]:
      arr[j],arr[j+1]=arr[j+1],arr[j]
print(arr)
```

```python
# Problem 2: Bubble Sort in Descending Order
# Problem Statement:
# Sort the array in descending order using Bubble Sort.
#
# Input:
# n
# array
#
# Output:
# Array in descending order
arr=list(map(int,input().split()))
for i in range(len(arr)):
  for j in range(len(arr)-i-1):
    if arr[j]<arr[j+1]:
      arr[j],arr[j+1]=arr[j+1],arr[j]
print(arr)
```

```python
# Problem 3: Count Bubble Sort Swaps
# Problem Statement:
# Count the number of swaps required to sort the array using Bubble Sort.
#
# Input:
# n
# array
#
# Output:
# Number of swaps
n=int(input())
arr=list(map(int,input().split()))

c=0
for i in range(n):
  for j in range(n-i-1):
    if arr[j]>arr[j+1]:
      c+=1
print(c)
```

```python
# Problem 4: Check Already Sorted
# Problem Statement:
# Check if the array is already sorted.
#
# Input:
# n
# array
#
# Output:
# YES if sorted, otherwise NO
n=int(input())
arr=list(map(int,input().split()))
flag=True
for i in range(n-1):
    if arr[i]>arr[i+1]:
      flag=False

if flag:
  print("Yes sorted")
else:
  print("No")
```

```python
# Problem 6: Selection Sort an Array
# Problem Statement:
# Sort an array using Selection Sort.
#
# Input:
# n
# array
#
# Output:
# Sorted array
def sortstring(arr):
  for i in range(len(arr)-1):
    min_index=i
    for j in range(i+1,len(arr)):
      if arr[j]<arr[min_index]:
        min_index=j
    arr[i],arr[min_index]=arr[min_index],arr[i]
  return arr

arr=list(map(int,input().split()))
print(sortstring(arr))
```

```python
# Problem 8: Selection Sort Descending
# Problem Statement:
# Sort the array in descending order using Selection Sort.
#
# Input:
# n
# array
#
# Output:
# Descending order array

n = int(input())
arr = list(map(int, input().split()))

for i in range(n-1):
    max_index = i

    for j in range(i + 1, n):
        if arr[j] > arr[max_index]:
            max_index = j

    arr[i], arr[max_index] = arr[max_index], arr[i]

print(*arr)
```

```python
# Problem 9: Kth Smallest Element
# Problem Statement:
# Find the Kth smallest element using Selection Sort logic.
#
# Input:
# n
# array
# k
#
# Output:
# Kth smallest element

n = int(input())
arr = list(map(int, input().split()))
k = int(input())

for i in range(k):
    min_index = i

    for j in range(i + 1, n):
        if arr[j] < arr[min_index]:
            min_index = j

    arr[i], arr[min_index] = arr[min_index], arr[i]

print(arr[k - 1])
```

```python
# Problem 10: Absolute Value Sorting
# Problem Statement:
# Sort the array based on absolute values.
#
# Input:
# n
# array
#
# Output:
# Array sorted by absolute value

n = int(input())
arr = list(map(int, input().split()))

for i in range(n-1):
    min_index = i

    for j in range(i + 1, n):
        if abs(arr[j]) < abs(arr[min_index]):
            min_index = j

    arr[i], arr[min_index] = arr[min_index], arr[i]

print(*arr)
```

```python
# Problem 11: Insertion Sort an Array
# Problem Statement:
# Sort an array using Insertion Sort.
#
# Input:
# n
# array
#
# Output:
# Sorted array

n = int(input())
arr = list(map(int, input().split()))

for i in range(1, n):
    key = arr[i]
    j = i - 1

    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        j -= 1

    arr[j + 1] = key

print(*arr)
```

```python
# Problem 12: Count Insertion Sort Shifts
# Problem Statement:
# Count the number of shifts performed during Insertion Sort.
#
# Input:
# n
# array
#
# Output:
# Number of shifts

n = int(input())
arr = list(map(int, input().split()))

count = 0

for i in range(1, n):
    key = arr[i]
    j = i - 1

    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        count += 1
        j -= 1

    arr[j + 1] = key

print(count)
```

```python
# Problem 13: Nearly Sorted Array
# Problem Statement:
# Sort a nearly sorted array using Insertion Sort.
#
# Input:
# n
# array
#
# Output:
# Sorted array

n = int(input())
arr = list(map(int, input().split()))

for i in range(1, n):
    key = arr[i]
    j = i - 1

    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        j -= 1

    arr[j + 1] = key

print(*arr)
```

```python
# Problem 14: Sort Characters
# Problem Statement:
# Sort the characters of a string using Insertion Sort.
#
# Input:
# string
#
# Output:
# Sorted string

arr = list(input())

for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1

    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        j -= 1

    arr[j + 1] = key

print("".join(arr))
```

```python
# Problem 15: Insert Element in Sorted Array
# Problem Statement:
# Insert an element into a sorted array using Insertion Sort logic.
#
# Input:
# n
# sorted array
# x
#
# Output:
# Updated sorted array

n = int(input())
arr = list(map(int, input().split()))
x = int(input())

arr.append(x)

i = len(arr) - 2

while i >= 0 and arr[i] > x:
    arr[i + 1] = arr[i]
    i -= 1

arr[i + 1] = x

print(*arr)
```
