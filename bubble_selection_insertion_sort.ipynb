```python
'''Bubble Sort Problem Statements:
Problem 1: Sort a list of strings in lexicographical (dictionary) order. (Level-Easy) '''
def bubble_sort_strings(arr):
  # write your code here
  for i in range(len(arr)-1):
    for j in range(0,len(arr)-i-1):
      if arr[j]>arr[j+1]:
        arr[j],arr[j+1]=arr[j+1],arr[j]
  return arr

my_strings = ["banana", "apple", "orange", "grape", "pear"]
bubble_sort_strings(my_strings)
print("Sorted strings:", my_strings)
```

```python
'''Problem 2: Sort a list of tuples based on a specific index of each tuple. (Level-Medium)'''
def bubble_sort_tuples(arr, index):
  # Write your code here
  for i in range(len(arr)-1):
    for j in range(0,len(arr)-i-1):
      if arr[j][index]>arr[j+1][index]:
        arr[j],arr[j+1]=arr[j+1],arr[j]
  return arr


my_tuples = [(3, 'apple'), (1, 'orange'), (2, 'banana'), (4, 'pear')]
bubble_sort_tuples(my_tuples, 0)
print("Sorted tuples based on the first element:", my_tuples)
```

```python
'''Problem 3: Sort a list of custom objects based on one of their attributes (Level-Medium) '''
class Person:
  def __init__(self,name,age):
    self.name=name
    self.age=age

def bubble_sort_objects(arr, attribute):
  #Write your code here
  for i in range(len(arr)-1):
    for j in range(0,len(arr)-i-1):
      if getattr(arr[j],attribute)>getattr(arr[j+1],attribute):
        arr[j],arr[j+1]=arr[j+1],arr[j]

people = [Person("John", 25), Person("Alice", 20), Person("Bob", 30)]
bubble_sort_objects(people, 'age')
print("Sorted people based on age:", [person.name for person in people])
```

```python
'''Insertion Sort Problem Statements:
Problem 1: Sort a list of strings in lexicographical (dictionary) order using insertion sort. (Level-Easy)'''
def insertion_sort_strings(arr):
  # write your code here
  for i in range(1,len(arr)):
    key=arr[i]
    j=i-1
    while j>=0 and key<arr[j]:
      arr[j+1]=arr[j]
      j-=1
    arr[j+1]=key


my_strings = ["banana", "apple", "orange", "grape", "pear"]
insertion_sort_strings(my_strings)
print("Sorted strings:", my_strings)
```

```python
'''Problem 3: Sort a list of tuples based on the sum of their elements using insertion sort. '''
def insertion_sort_tuples_by_sum(arr):
  # write your code here
  for i in range(1,len(arr)):
    key=arr[i]
    j=i-1
    while j>=0 and sum(key)<sum(arr[j]):
      arr[j+1]=arr[j]
      j-=1
    arr[j+1]=key
my_tuples = [(3, 7), (1, 5), (2, 4), (4, 6)]
insertion_sort_tuples_by_sum(my_tuples)
print("Sorted tuples by sum of elements:", my_tuples)
```
