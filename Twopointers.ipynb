```python
'''Example 1: Find Pair with Given Sum
Problem Given a sorted array and a target sum,
find whether a pair exists.
arr = [1, 2, 3, 4, 6, 8] target = 10 '''
arr=[1,2,3,4,5,6,8]
target=10

left=0
right=len(arr)-1

while left<right:
  sum=arr[left]+arr[right]

  if sum==target:
    print(arr[left],arr[right])
    break

  elif sum<target:
    left+=1

  else:
    right-=1
```

```python
'''Example 2: Reverse an Array '''
arr= [10, 20, 30, 40, 50]
l=0
r=len(arr)-1

while l<r:
  arr[l],arr[r]=arr[r],arr[l]
  l+=1
  r-=1
print(arr)
```
