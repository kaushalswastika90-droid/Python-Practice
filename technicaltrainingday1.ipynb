```python
n = int(input("Enter a number: "))

i = 2
is_prime = True

while i < n:
    if n % i == 0:
        is_prime = False
        break
    i += 1

if n <= 1:
    print("Not Prime")
elif is_prime:
    print("Prime Number")
else:
    print("Not Prime")
```

```python
n = 5

i = n
while i >= 1:
    j = n
    while j >= i:
        print("*", end=" ")
        j -= 1
    print()
    i -= 1
```

```python
n = 5

i = n
while i >= 1:
    j = n
    while j >= n - i + 1:
        print(j, end=" ")
        j -= 1
    print()
    i -= 1
```
