```python
"""3. Count Vowels and Consonants
Input: A string s
Output: Two integers: vowel count and consonant count
Sample Input: hello
Sample Output: 2 3"""

s = input("Enter a string: ")
vowels = "aeiouAEIOU"

vowel_count = sum(1 for char in s if char in vowels)
consonant_count = sum(1 for char in s if char.isalpha() and char not in vowels)

print(vowel_count, consonant_count)
```

```python
"""7. Character Frequency
Input: A string s
Output: Frequency of each character sorted
Sample Input: banana
Sample Output: a 3 b 1 n 2"""
a=input("enter a string: ")
b=""

for ch in a:
  if ch not in b:
    print(ch,a.count(ch))
    b+=ch

```

```python
"""8. First Non-Repeating Character
Input: A string s
Output: First non-repeating character or -1
Sample Input: aabbccde
Sample Output: d"""

a = input("enter a string: ")

def check(s):
    for ch in s:
        if s.count(ch) == 1:
            return ch
    return -1

print(check(a))
```

```python
"""9. Check Subsequence
Input: Two strings s1 and s2
Output: YES if s1 is subsequence of s2 else NO
Sample Input: abc axbycz
Sample Output: YES"""

s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

i=0
j=0

while i<len(s1) and j<len(s2):
    if s1[i] == s2[j]:
        i += 1
    j += 1

if i == len(s1):
    print("YES")
else:
    print("NO")
```

```python
"""10. String Compression (Run-Length Encoding)
Input: A string s
Output: Compressed string
Sample Input: aaabbc
Sample Output: a3b2c1"""
a = input("enter a string: ")
count = 1

for i in range(len(a)):
    if i < len(a)-1 and a[i] == a[i+1]:
        count += 1
    else:
        print(a[i] + str(count), end="")
        count = 1
```

```python
"""13. Check Rotation
Input: Two strings s1 and s2
Output: YES if rotation else NO
Sample Input: abcde deabc
Sample Output: YES"""
s1=input("Enter s1: ")
s2=input("Enter s2: ")

if len(s1)!=len(s2):
  print("NO")

else:
  temp=s1+s1
  if s2 in temp:
    print("YES")
  else:
    print("NO")
```

```python
"""14. Pangram Check
Input: A sentence
Output: YES if pangram else NO
Sample Input: The quick brown fox jumps over a lazy dog
Sample Output: YES"""

s=input("enter string: ").lower()

alphabet="abcdefghijklmnopqrstuvwxyz"

for i in alphabet:
  if i not in s:
    print("No")
    break
else:
  print("Yes")
```

```python
"""18. Remove All Digits
Input: A string s
Output: String without digits
Sample Input: pyth0n123
Sample Output: pythn"""

s=input("enter a string: ")
result=""

for ch in s:
  if ch.isalpha():
    result+=ch

print(result)
```
