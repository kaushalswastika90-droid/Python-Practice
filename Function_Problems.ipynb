```python
from functools import reduce

# 1. Sum of Two Numbers
def sum_two(a, b):
    return a + b

# 2. Maximum of Three Numbers
def max_three(a, b, c):
    return max(a, b, c)

# 3. Multiply All Numbers using *args
def multiply_all(*args):
    result = 1
    for i in args:
        result *= i
    return result

# 4. Concatenate Strings using *args
def concat_strings(*args):
    return "".join(args)

# 5. Print Info using **kwargs
def print_info(**kwargs):
    for k, v in kwargs.items():
        print(f"{k}: {v}")

# 6. Lambda Square
square = lambda x: x * x

# 7. Map Square
def map_square(lst):
    return list(map(lambda x: x*x, lst))

# 8. Filter Even
def filter_even(lst):
    return list(filter(lambda x: x % 2 == 0, lst))

# 9. Filter Strings Length > 3
def filter_long_words(lst):
    return list(filter(lambda x: len(x) > 3, lst))

# 10. Reduce Product
def reduce_product(lst):
    return reduce(lambda x, y: x*y, lst)

# 11. Count Vowels
def count_vowels(s):
    vowels = "aeiouAEIOU"
    return sum(1 for ch in s if ch in vowels)

# 12. Reverse Words
def reverse_words(s):
    return " ".join(s.split()[::-1])

# 13. Palindrome Check
def is_palindrome(s):
    return "Palindrome" if s == s[::-1] else "Not Palindrome"

# 14. Minimum Value
def find_min(lst):
    return min(lst)

# 15. Map Convert to String
def convert_to_string(lst):
    return list(map(str, lst))

# 16. Filter Numbers > 10
def greater_than_10(lst):
    return list(filter(lambda x: x > 10, lst))

# 17. Sort by Last Digit
def sort_by_last_digit(lst):
    return sorted(lst, key=lambda x: x % 10)

# 18. Sum using *args
def sum_all(*args):
    return sum(args)

# 19. Marksheet using **kwargs
def marksheet(**kwargs):
    for subject, marks in kwargs.items():
        print(f"{subject}: {marks}")

# 20. Reduce Maximum
def reduce_max(lst):
    return reduce(lambda x, y: x if x > y else y, lst)


# ---- Example Calls ----
print(sum_two(4,5))
print(max_three(10,5,18))
print(multiply_all(1,2,3,4))
print(concat_strings("Hello","World","Python"))
print_info(name="Aniket", age=22, city="Mumbai")
print(square(5))
print(map_square([1,2,3,4,5]))
print(filter_even([1,2,3,4,5,6]))
print(filter_long_words(["cat","dog","lion","tiger"]))
print(reduce_product([1,2,3,4]))
print(count_vowels("hello world"))
print(reverse_words("hello world python"))
print(is_palindrome("madam"))
print(find_min([10,4,30,2,7]))
print(convert_to_string([1,2,3,4]))
print(greater_than_10([5,12,7,20,3,25]))
print(sort_by_last_digit([23,91,34,72]))
print(sum_all(5,10,15,20))
marksheet(math=90, sci=88, eng=92)
print(reduce_max([6,9,3,12,5]))
```
