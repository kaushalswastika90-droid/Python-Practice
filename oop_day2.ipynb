```python
'''polymorphism-One name, many forms.'''

class India:
    def capital(self):
        print("Capital: New Delhi")

    def language(self):
        print("Language: Hindi")

    def type(self):
        print("Type: Developing Country")


class USA:
    def capital(self):
        print("Capital: Washington D.C.")

    def language(self):
        print("Language: English")

    def type(self):
        print("Type: Developed Country")


countries = [India(), USA()]

for country in countries:
    country.capital()
    country.language()
    country.type()
    print()
```

```python
class Person:
    def __init__(self, name):
        self.name = name

    def display(self):
        print("Name:", self.name)


class Student(Person):
    def __init__(self, name, roll):
        super().__init__(name)
        self.roll = roll

    def display(self):   # method overriding (polymorphism)
        super().display()
        print("Roll no:", self.roll)


# Creating object
s = Student("Swastika", 101)
s.display()
```

```python
"""Duck Typing is a Python concept where an object's suitability is determined by the presence of required methods and attributes
rather than its actual class or type. The focus is on behavior, not inheritance."""
class Duck:
    def sound(self):
        print("Quack Quack")

class Dog:
    def sound(self):
        print("Bow Bow")

def make_sound(animal):
    animal.sound()

d = Duck()
dog = Dog()

make_sound(d)     # Quack Quack
make_sound(dog)   # Bow Bow
```

```python
from os import name
"""Problem 1: Single Inheritance – Person and Student Problem Statement: Create a Person class with attributes name and age.
Inherit a Student class from Person with an additional attribute roll_no.
Print the details of a student.
Input Format: First line: Name (string), Second line: Age (integer), Third line: Roll number (integer)
Output Format: Print name, age, and roll number in separate lines.
Sample Input: Alice | 20 | 101
Sample Output: Name: Alice | Age: 20 | Roll No: 101"""

class Person:
  def __init__(self,name,age):
    self.name=name
    self.age=age

  def display(self):
        print("Name:", self.name)
        print("Age:", self.age)

class Student(Person):
  def __init__(self, name,age, roll):
        super().__init__(name,age)
        self.roll = roll

  def display(self):
        super().display()
        print("Roll no:", self.roll)

name=input()
age=int(input())
roll=int(input())
s=Student(name,age,roll)
s.display()

```

```python
"""Problem 2: Multilevel Inheritance – Vehicle and Car
Problem Statement: Create a Vehicle class with attribute brand. Create a Car class inheriting Vehicle with attribute model.
Create a SportsCar class inheriting Car with attribute top_speed.
Print details of a SportsCar.
Input Format:
First line: brand (string),
Second line: model (string),
Third line: top speed (integer)
Output Format: Print brand, model, and top speed.
Sample Input: Ferrari | F8 | 340
Sample Output: Brand: Ferrari | Model: F8 | Top Speed: 340"""
class Vehicle:
  def __init__(self,brand):
    self.brand=brand

  def display(self):
    print("brand: ",self.brand)

class Car(Vehicle):
  def __init__(self,brand,model):
    super().__init__(brand)
    self.model=model

  def display(self):
    super().display()
    print("model: ",self.model)

class SportsCar(Car):
  def __init__(self,brand,model,top_speed):
    super().__init__(brand,model)
    self.top_speed=top_speed

  def display(self):
    super().display()
    print("top speed: ",self.top_speed)

brand=input()
model=input()
top_speed=int(input())
s=SportsCar(brand,model,top_speed)
s.display()
```

```python
"""Problem 3: Hierarchical Inheritance – Teacher and Student Problem Statement: Create a base class Person with attribute name.
Derive two classes: Teacher with attribute subject, Student with attribute grade. Print details.
Input Format: Teacher name, subject, student name, grade
Output Format: Teacher details and student details
Sample Input: Mr. Smith | Math | Alice | A
Sample Output: Teacher Name: Mr. Smith, Subject: Math | Student Name: Alice, Grade: A"""

class Person:
  def __init__(self,name):
    self.name=name

  def display(self):
    print("name",self.name)

class Teacher(Person):
  def __init__(self,name,subject):
    super().__init__(name)
    self.subject=subject

  def display(self):
    super().display()
    print("tname: ",self.name,"subject: ",self.subject)

class Student(Person):
  def __init__(self,name,grade):
    super().__init__(name)
    self.grade=grade

  def display(self):
    super().display()
    print("sname: ",self.name,"grade",self.grade)

tname=input()
subject=input()
sname=input()
grade=input()
t=Teacher(tname,subject)
s=Student(sname,grade)
t.display()
s.display()
```

```python
"""Problem 4: Inheritance with Method OverridingProblem Statement: Create a Shape class with method area() returning 0.
Derive Rectangle class with length and width, override area() to return rectangle area.
Input Format: Two integers length and width
Output Format: Print area Sample Input: 5 | 4 Sample Output: 20"""
class Shape:
  def area(self):
    return 0

class Rectangle(Shape):
  def __init__(self,length,width):
    self.length=length
    self.width=width
  def area(self):
    return self.length*self.width

length=int(input())
width=int(input())
r=Rectangle(length,width)
print(r.area())
```

```python
"""Problem 5: Super Keyword in Inheritance Problem Statement: Create Person class with constructor for name.
Create Employee class inheriting Person with constructor for salary using super().
Print name and salary.
Input Format: name, salary
Output Format: Print name and salary
Sample Input: Alice | 50000
Sample Output: Name: Alice | Salary: 50000"""
class Person:
  def __init__(self,name):
    self.name=name

  def display(self):
    print("name: ",self.name)

class Employee(Person):
  def __init__(self,name,salary):
    super().__init__(name)
    self.salary=salary

  def display(self):
    super().display()
    print("salary: ",self.salary)

name=input()
salary=int(input())
e=Employee(name,salary)
e.display()
```

```python
"""Problem 7: Duck Typing Problem Statement: Write Car and Bike classes both with wheels() method.
Create function display_vehicle(obj).
Input Format: Car or Bike
Output Format: Print wheels info
Sample Input: Bike
Sample Output: Two Wheels"""

class car:
  def wheels(self):
    print("Four Wheels")

class bike:
  def wheels(self):
    print("Two Wheels")

def display_vehicle(obj):
  obj.wheels()

type=input()
if type=="car":
  c=car()
  display_vehicle(c)
elif type=="bike":
  b=bike()
  display_vehicle(b)
else:
  print("invalid input")
```

```python
"""Problem 8: Operator Overloading
Problem Statement: Create Vector class with x, y. Overload + to add vectors.
Input Format: x1 y1 | x2 y2
Output Format: (x, y)
Sample Input: 2 3 | 4 5
Sample Output: (6, 8)"""
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):   # overload +
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):          # for output format
        return f"({self.x}, {self.y})"



x1, y1 = map(int, input().split())
x2, y2 = map(int, input().split())


v1 = Vector(x1, y1)
v2 = Vector(x2, y2)

result = v1 + v2

print(result)
```

```python
"""Problem 9: Method Overloading Simulation Problem Statement:
Create Calculator class with add() method using default args to accept 2 or 3 numbers.
Input Format: count n1 n2 (n3)
Output Format: Sum
Sample Input: 3 | 5 7 9
Sample Output: 21"""
class calculator:
  def add(self,a,b,c=None):
    if c==None:
      return a+b
    else:
      return a+b+c

count = int(input())

nums = list(map(int, input().split()))

calc = calculator()

if count == 2:
    print(calc.add(nums[0], nums[1]))
elif count == 3:
    print(calc.add(nums[0], nums[1], nums[2]))
```

```python
"""Problem 10: Abstract Class Polymorphism Problem Statement: Create abstract Shape with area().
Implement Rectangle and Circle.
Input Format: Shape type and parameters
Output Format: Print area
Sample Input: Rectangle | 5 | 4
Sample Output: 20
Sample Input: Circle | 7
Sample Output: 153"""

class Shape:
  def area(self):
    return 0

class Rectangle(Shape):
  def __init__(self,length,width):
    self.length=length
    self.width=width

  def area(self):
    return self.length*self.width

class Circle(Shape):
  def __init__(self,radius):
    self.radius=radius

  def area(self):
    return 3.14*self.radius*self.radius

type=input()
if type=="Rectangle":
  length=int(input())
  width=int(input())
  r=Rectangle(length,width)
  print(r.area())
elif type=="Circle":
  radius=int(input())
  c=Circle()
  print(c.area(radius))
else:
  print("Invalid input")
```

```python
class ComplexNumber:
    def __init__(self, x, y):   # constructor
        self.x = x   # real part
        self.y = y   # imaginary part

    def __add__(self, other):   # operator overloading
        real = self.x + other.x
        imag = self.y + other.y
        return ComplexNumber(real, imag)

    def __str__(self):          # for printing
        return f"{self.x} + {self.y}i"


# Objects
c1 = ComplexNumber(2, 3)
c2 = ComplexNumber(4, 5)

# Addition
result = c1 + c2

print(result)
```

```python
class Bank:
  account_number_generator=2024080300
  def __init__(self, name, gender, amount):
          self.name = name
          self.gender = gender
          self.amount = amount

          Bank.account_number_generator += 1
          self.account_no = Bank.account_number_generator

          print("Congrats", self.name, "Account created and your account number is", self.account_no)

  def withdraw(self, amount):
        if self.amount - amount >= 3000:
            self.amount -= amount
            print("Withdrawal successful. Remaining balance:", self.amount)
        else:
            print("Transaction denied: minimum balance must be 3000")

  def deposit(self,amount):
    if amount>0:
      self.amount += amount
      print("account no: ", self.account_no,"Deposited successfully. New balance:", self.amount)
    else:
      print("account cannot be -ve")

  def check_balance(self):
    print("account no: ",self.account_no,"on name of:", self.name,"Balance: ",self.amount)

b1 = Bank("Aman", "M", 5000)

b1.deposit(2000)
b1.withdraw(3000)
b1.check_balance()
```
