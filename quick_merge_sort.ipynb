```python
'''Question 1: Student Marks Ranking (Quick Sort)
Problem Statement

A teacher wants to rank students based on their exam marks. Given the marks of n students, sort the marks in ascending order using Quick Sort.

Input Format
n
Marks array
Output Format
Sorted marks
Example Input
5
78 45 90 60 72
Example Output
45 60 72 78 90'''
def partition(arr, low, high):
    i = (low - 1)
    pivot = arr[high]

    for j in range(low,high):
      if arr[j]<pivot:
        i+=1
        arr[i],arr[j]=arr[j],arr[i]
    arr[i+1],arr[high]=arr[high],arr[i+1]
    return i+1

def qs(arr,low,high):
  if low<high:
    pi=partition(arr,low,high)
    qs(arr,low,pi-1)
    qs(arr,pi+1,high)

n=int(input())
marks=list(map(int,input().split()))
qs(marks,0,n-1)
print(*marks)
```

```python
'''Problem Statement

Find the median patient waiting time after sorting the waiting times using Quick Sort.

Input Format
n
Waiting times
Output Format
Median waiting time
Example Input
7
15 30 10 20 25 5 40
Example Output
20'''
def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1

    for j in range(low, high):
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def quick_sort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi + 1, high)

n = int(input())
waiting = list(map(int, input().split()))

quick_sort(waiting, 0, n - 1)

if n % 2 == 1:
    print(waiting[n // 2])
else:
    median = (waiting[n // 2 - 1] + waiting[n // 2]) / 2
    print(median)
```

```python
'''Question 2: Fastest Delivery Time (Quick Sort – Kth Smallest)
Problem Statement

A delivery company records delivery times. Find the Kth fastest delivery time using Quick Sort partition logic.

Input Format
n
Delivery times
k
Output Format
Kth fastest time
Example Input
6
45 30 60 20 50 40
3
Example Output
40'''
def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1

    for j in range(low, high):
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def quick_sort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi + 1, high)

n = int(input())
times = list(map(int, input().split()))
k = int(input())

quick_sort(times, 0, n - 1)

print(times[k - 1])
```

```python
'''Question 3: Online Order Sorting (Merge Sort)
Problem Statement

Sort online order amounts using Merge Sort.

Input Format
n
Order amounts
Output Format
Sorted order amounts
Example Input
6
1200 500 1500 700 300 1000
Example Output
300 500 700 1000 1200 1500'''
def ms(arr):
  if len(arr)>1:
    mid=len(arr)//2
    left=arr[:mid]
    right=arr[mid:]

    ms(left)
    ms(right)

    i=j=k=0

    while i<len(left) and j<len(right):
      if left[i]<right[j]:
        arr[k]=left[i]
        i+=1
      else:
        arr[k]=right[j]
        j+=1
      k+=1

    while i<len(left):
      arr[k]=left[i]
      i+=1
      k+=1

    while j<len(right):
      arr[k]=right[j]
      j+=1
      k+=1

n=int(input())
orders=list(map(int,input().split()))
ms(orders)
print(*orders)
```

```python
'''Question 5: Sorting Files by Size (Merge Sort)
Problem Statement

Sort file sizes to optimize storage using Merge Sort.

Input Format
n
File sizes
Output Format
Sorted file sizes
Example Input
4
1200 300 800 500
Example Output
300 500 800 1200'''
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left = arr[:mid]
        right = arr[mid:]

        merge_sort(left)
        merge_sort(right)

        i = j = k = 0

        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                arr[k] = left[i]
                i += 1
            else:
                arr[k] = right[j]
                j += 1
            k += 1

        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1

n = int(input())
files = list(map(int, input().split()))

merge_sort(files)

print(*files)
```

```python
'''Question 4: Traffic Violation Analysis (Merge Sort – Inversion Count)
Problem Statement

Count violations where faster vehicles appear before slower ones using Merge Sort.

Input Format
n
Vehicle speeds
Output Format
Number of violations
Example Input
5
80 60 70 50 90
Example Output
5'''
def merge(arr):
    if len(arr) <= 1:
        return arr, 0

    mid = len(arr) // 2
    left, inv_left = merge(arr[:mid])
    right, inv_right = merge(arr[mid:])

    result = []
    i = j = 0
    inv = inv_left + inv_right

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            inv += len(left) - i
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])

    return result, inv

n = int(input())
speeds = list(map(int, input().split()))

_, count = merge(speeds)

print(count)
```
