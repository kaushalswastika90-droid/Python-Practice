```python
"""1. Reverse a List
Problem Statement: Given a list of integers, reverse the list.
Input Format:
- First line: integer n (size of list)
- Second line: n integers separated by space
Output Format:
- Print the reversed list as space-separated integers."""
n=int(input())
l=list(map(int,input().split()))
print(*l[::-1])
```

```python
"""2. Maximum Element in a List
Problem Statement: Find the maximum element in the list.
Input Format:
- First line: integer n
- Second line: n integers
Output Format:
- Single integer (maximum element)."""
n=int(input())
list1=list(map(int,input().split()))
print(max(list1))
```

```python
"""3. Count Even and Odd Numbers
Problem Statement: Count how many even and odd numbers are in the list.
Input Format:
- First line: integer n
- Second line: n integers
Output Format:
- Two integers separated by space → (count_even count_odd)"""
n=int(input())
list1=list(map(int,input().split()))
count_even=0
count_odd=0
for x in list1:
  if x%2==0:
    count_even+=1
  else:
    count_odd+=1
print("odd:",count_odd)
print("even:",count_even)
```

```python
"""4. Remove Duplicates
Problem Statement: Remove duplicate elements from the list while preserving order.
Input Format:
- First line: integer n
- Second line: n integers
Output Format:
- List without duplicates (space-separated)."""
n=int(input())
list1=list(map(int,input().split()))
d=[]
for x in list1:
  if x not in d:
    d.append(x)
  else:
    pass
print(*d)
```

```python
"""5. Second Largest Element
Problem Statement: Find the second largest number in the list.
Input Format:
- First line: integer n
- Second line: n integers
Output Format:
- Single integer (second largest element)."""
n=int(input())
list1=list(map(int,input().split()))
a=list(set(list1))
a.sort()
print(a[-2])
```

```python
"""6. Rotate List
Problem Statement: Rotate a list by k positions to the right.
Input Format:
- First line: integer n (size of list)
- Second line: n integers
- Third line: integer k
Output Format:
- Rotated list as space-separated integers."""
n=int(input())
l1=list(map(int,input().split()))
k=int(input())
k=k%n
lst=l1[-k:]+l1[:-k]
print(*lst)
```

```python
"""8. Merge Two Lists
Problem Statement: Merge two lists and print the sorted merged list.
Input Format:
- First line: integer n1
- Second line: n1 integers
- Third line: integer n2
- Fourth line: n2 integers
Output Format:
- Sorted merged list (space-separated)."""
n1=int(input())
l1=list(map(int,input().split()))
n2=int(input())
l2=list(map(int,input().split()))
l3=l1+l2
l3.sort()
print(*l3)
```

```python
"""9. Palindrome List
Problem Statement: Check if the list is a palindrome.
Input Format:
- First line: integer n
- Second line: n integers
Output Format:
- Print YES if palindrome, else NO."""
n=int(input())
l1=list(map(int,input().split()))
if l1==l1[::-1]:
  print("YES")
else:
  print("NO")
```

```python
"""10. List Comprehension – Squares
Problem Statement: Given a number n, generate a list of squares of numbers from 1 to n.
Input Format:
- Single integer n
Output Format:
- List of squares (space-separated)."""
n=int(input())
s=[i*i for i in range(1,n+1)]
print(*s)
```

```python
"""14. Subset Check
Problem Statement: Check if set A is a subset of set B.
Input Format:
- First line: integer n1 followed by n1 integers
- Second line: integer n2 followed by n2 integers
Output Format:
- Print YES if subset, else NO."""
n1 = int(input())
a = list(map(int, input().split()))

n2 = int(input())
b = list(map(int, input().split()))

if set(a).issubset(set(b)):
    print("YES")
else:
    print("NO")
```

```python
"""16. Frequency of Elements
Problem Statement: Count the frequency of each element in the list using a dictionary.
Input Format:
- First line: integer n
- Second line: n integers
Output Format:
- Each element and its frequency in separate lines."""
n = int(input())
lst = list(map(int, input().split()))

freq = {}
for x in lst:
    freq[x] = freq.get(x, 0) + 1

for k, v in freq.items():
    print(k, v)
```

```python
"""18. Word Count
Problem Statement: Count the number of occurrences of each word in a sentence.
Input Format:
- Single line: string (sentence)
Output Format:
- Each word and its frequency."""
s = input().split()

freq = {}
for word in s:
    freq[word] = freq.get(word, 0) + 1

for k, v in freq.items():
    print(k, v)
```

```python
"""19. Student Marks Dictionary
Problem Statement: Create a dictionary where keys are student names and values are their marks.
Given a name, print the marks.
Input Format:
- First line: integer n (number of students)
- Next n lines: name mark
- Last line: query name
Output Format:
- Print the marks of the queried student."""
n = int(input())

d = {}
for _ in range(n):
    name, marks = input().split()
    d[name] = int(marks)

query = input()
print(d.get(query, "Not found"))
```

```python
"""20. Merge Two Dictionaries
Problem Statement: Merge two dictionaries into one. If a key exists in both, add their values.
Input Format:
- First line: integer n1 (number of key-value pairs in dict1)
- Next n1 lines: key value
- Next line: integer n2
- Next n2 lines: key value
Output Format:
- Print merged dictionary (key value on each line)"""
n1 = int(input())
d1 = {}

for _ in range(n1):
    k, v = input().split()
    d1[k] = int(v)

n2 = int(input())
d2 = {}

for _ in range(n2):
    k, v = input().split()
    d2[k] = int(v)

# merge
for k in d2:
    d1[k] = d1.get(k, 0) + d2[k]

for k, v in d1.items():
    print(k, v)
```
